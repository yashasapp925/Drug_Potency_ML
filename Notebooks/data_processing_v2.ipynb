```python
# Imports

import pandas as pd
from chembl_webresource_client.new_client import new_client
import numpy as np
```

```python
# Use ChEMBL Python API to search for a disease or protein of interest

target = new_client.target
target_query = target.search('coronavirus')
targets_df = pd.DataFrame.from_dict(target_query)
targets_df
```

```python
# Select desired target

selected_target = targets_df['target_chembl_id'][6]
selected_target
```

```python
# Retrieve bioactivity data reported as IC50 values

activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = 'IC50')
res_df = pd.DataFrame.from_dict(res)
res_df
```

```python
res_df.to_csv('Datasets/sars_cov_bioactivity_data.csv', index=False)
```

```python
# Handle missing data

res_df = res_df[res_df['standard_value'].notna()]
res_df
```

```python
# Label compounds as either active, inactive, or intermediate

standard_values = res_df['standard_value'].astype(float).values
bioactivity_class = np.empty(standard_values.shape, dtype='<U12')

bioactivity_class[standard_values >= 10000] = 'inactive'
bioactivity_class[standard_values <= 1000] = 'active'
bioactivity_class[(standard_values > 1000) & (standard_values < 10000)] = 'intermediate'
```

```python
# Extract necessary columns for analysis

combined_df = res_df[['molecule_chembl_id', 'canonical_smiles', 'standard_value']]
combined_df['bioactivity_class'] = bioactivity_class
```

```python
combined_df
```

```python
# Convert IC50 standard values to negative logarithmic scale for more uniform distribution

final_df = combined_df.copy()
norm_std_val = final_df['standard_value'].astype(float).clip(upper=100000000)
final_df['pIC50'] =  -np.log10(norm_std_val * 1e-9)
final_df.drop(columns=['standard_value'], inplace=True)
final_df
```

```python
# Remove intermediate class samples for simplicity

final_df = final_df[final_df['bioactivity_class'] != 'intermediate']
final_df
```

```python
final_df.to_csv('Datasets/sars_cov_bioactivity_preprocessed_data.csv')
```
